```python
import yfinance as yf
from pydantic import BaseModel

class OpportunityState(BaseModel):
    symbol: str
    prices: dict | None = None
    indicators: dict | None = None
    score: float | None = None
    action: str | None = None


def fetch_data(state):
    ticker = yf.Ticker(state.symbol)
    last_price = ticker.history(period="1d")["Close"].iloc[-1]

    state.indicators = {
        "above_sma20": last_price  > 20,
        "value_zone": last_price  < 20
    }


    state.prices = {"last_close": float(last_price).__round__(2)}
    return state

price = fetch_data(OpportunityState(symbol="AMZN"))
print(price)
```
